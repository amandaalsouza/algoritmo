# Rodando a sua primeira IA no Google Colab


```python
import os
from google.colab import userdata

os.environ['GOOGLE_API_KEY'] = userdata.get('GEMINI_PROJETO1_KEY')
```

```python
from google import genai

client = genai.Client()
```

```python
resposta = client.models.generate_content(model='gemini-2.5-flash', contents='o que é a inteligencia artificial?')
```

```python
print(resposta.text)
```

# Meu primeiro Chatbot

```python
chatbot = client.chats.create(model='gemini-2.5-flash')
```

```python
resposta = chatbot.send_message('o que é a inteligencia artificial?')
```

```python
print(resposta.text)
```

```python
resposta = chatbot.send_message('o que é a lua?')
print(resposta.text)
```

```python
chatbot.get_history()
```

# Python e inteligencia artificial

```python
prompt = input('Digite a sua pergunta:')

while prompt != 'sair':
  resposta = chatbot.send_message(prompt)
  print(resposta.text)
  print('\n')
  prompt = input('Digite a sua pergunta:')
```
